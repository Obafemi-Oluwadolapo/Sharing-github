```python
#To get the heatmap for each of the variables
correlation_mat = abs(df.corr())
top_correlation = correlation_mat.index
plt.figure(figsize=(20,20))
#plotting the heat map for the variables
e = snsport.heatmap(abs(df[top_correlation].corr()),annot=True,cmap="RdYlGn")
```

The Exploratory Data Analysis using the heat map, it was seen from variables which leads us to features that holds great importance. These features were **cp**, **thalach**, **exang** and **oldpeak**. This implies that **people with heart diseases** will most likely have **chest pain, maximum heart rate, exercise induced anginahh,** and **high ST depression** induced by **exercise relative to rest**.

```python
e = snsport.pairplot(df, height=3, vars=["cp", "thalach",'exang','oldpeak','target'])
```

## The result better confirms the result gotten from the pairplot

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#splitting into train and test data
X = df[['thalach','oldpeak']]
y = df['target']
X_t, X_test, y_t, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
lf = RandomForestClassifier(n_estimators=1000,random_state=100)
lf.fit(X,y)
pred = lf.predict(X_test)

# accuracy of the dataset
print(metrics.accuracy_score(y_test,pred))
```
